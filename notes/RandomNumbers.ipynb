```julia
using Plots, Distributions
```

```julia
m = 1249
function gen(prev)
    a = 421
    c = 821
    return (a*prev+c)%m
end
prev = 1
data = []
n = 2000
for i in 1:n
    prev = gen(prev)
    push!(data, prev)
end
vals = zeros(m)
for i in data
    vals[i+1] += 1
end
plot(seriestype=:scatter, vals)
```
